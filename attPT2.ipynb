```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Carregar o dataset
df = pd.read_csv('aluguel_duas_variaveis_com_ruido.csv')

# Explora√ß√£o dos dados
print("Primeiras linhas:\n")
print(df.head())
print("\nEstat√≠sticas:")
print(df.describe())

# Visualiza√ß√£o da rela√ß√£o entre as vari√°veis
plt.figure(figsize=(12, 6))

# Gr√°fico de dispers√£o para a rela√ß√£o entre √Årea e Pre√ßo do Aluguel
plt.subplot(1, 2, 1)
plt.scatter(df['Area'], df['Preco'])
plt.xlabel('√Årea (m¬≤)')
plt.ylabel('Pre√ßo do Aluguel (R$)')
plt.title('Pre√ßo do Aluguel x √Årea da Casa')

# Gr√°fico de dispers√£o para a rela√ß√£o entre Quartos e Pre√ßo do Aluguel
plt.subplot(1, 2, 2)
plt.scatter(df['Quartos'], df['Preco'])
plt.xlabel('N√∫mero de Quartos')
plt.ylabel('Pre√ßo do Aluguel (R$)')
plt.title('Pre√ßo do Aluguel x N√∫mero de Quartos')

plt.tight_layout()
plt.show()

# Prepara√ß√£o dos dados
X = df[['Area', 'Quartos']]
y = df['Preco']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Constru√ß√£o do modelo
model = LinearRegression()
model.fit(X_train, y_train)

# Avalia√ß√£o do modelo
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f"R¬≤: {r2:.2f}")
```
